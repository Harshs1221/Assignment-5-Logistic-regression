```python
#Q5.  What is Regularization in Logistic Regression? Why is it needed?
"""
Ans5. Regularization in Logistic Regression is a technique used to prevent overfitting by adding a penalty to the
      model's complexity.
      It ensures the model generalizes well to unseen data instaed of memorizing the training data.

      Why is it needed?
      - Reduces Overfitting(L2 regularization (ridge)):
              Prevents the model from learning noise and irrelevant patterns.
      - Feature Selection (L1 regularization (lasso)):
              Helps eliminate less useful features by shrinking their coefficients.
      - Improves Generalization:
              Ensures better performance on new, unseen data.
"""
```

```python
#Q6.  Explain the difference between Lasso, Ridge, and Elastic Net regression?
"""
Ans6. Lasso Regression (L1 Regularization):
        - Adds the absolute values of coefficients as a penalty.
        - Some coefficients shrink to zero, effectively removing less important features.
        - Useful when you want automatic feature selection.

      Ridge regression (L2 Regularization)
        - Adds the squared values of coefficients as a penalty.
        - Reduces coefficient magnitides biy does not eliminate features.
        - Helps when features are highly correlated.

      Elastic Net Regression
        - Combines L1 and L2 regularization.
        - Useful when there are many correlated features.
        - Provide a balance between feature selection (Lasso) and coefficient shrinkage (Ridge).
"""
```

```python
#Q7.  When should we use Elastic Net instead of Lasso or Ridge?
"""
Ans7. Elastic net is used when you want a balance between feature selection and coefficient shrinkage.
      When to use Elastic net:
        When Lasso is removing too many features - Elastic net keeps the important ones.
        When features are highly correlated- It balances between feature selection and keeping useful features.
        When Ridge alone is too weak in reducing complexity - Elastic net provides a stringer regularization.
"""
```

```python
#Q12.  How does class imbalance affect Logistic Regression?
"""
Ans12. Class imbalance can significantly impact logistic regression it will give biased predictions, If one class has
       way more data than the other, the model may favour it and ignore the smaller class.
       2. Skewed Probability Estimates :
               The predicted probabilities may not be reliable, as the model learns patterns mostly from the dominant class.

        Effect:
            The model might have high accuracy overall but struggle to correctly predict the minority class.
        Solution:
            You can balance the data by adjusting class weights, using better evaluation metrics(like precision & recall),
            or applying techniques like oversampling / undersampling.
"""
```

```python
#Q14. What are different solvers in Logistic Regression? Which one should be used?
"""
Ans14. Logistic Regression offers several solvers:
        1. Liblinear -
                Best for small to medium-sized datasets and supports L1 and L2 regularization.
        2. LBFGS -
                Works well for large datasets and supports L2 regularization.
        3.Newton-CG
                Suitable for datasets with many features, but mainly supports L2 regularization.
        4. SAG (Stochastic Average Gradient) -
                Efficient for large datasets but require L2 regularization.
        5. SAGA -
                Similar to SAG but supports both L1 and L2 regularization, making it useful for sparse data.
         Which one to use:
         1. If your dataset is small, use liblinear.
         2. If your dataset is large, use LBFGS, SAG, pr SAGA
         3. If you have many Fatures, try newton-CG
         4. If your data is sparse, SAGA is a good choice.
"""
```

```python
#Q15.  How is Logistic Regression extended for multiclass classification?
"""
Ans15. Logistic Regression is usually used for two class problem (like Yes/No), but it can be extended for
       multiclass classification using these methods:

       1. One-vs-Rest (OvR) -
               The model creates separate classifiers for each class and picks the one with the highest probability.
       2. One-vs-One (OvO) -
                The model builds classifiers for every pair of classes and decides based on majority voting.
       3. Multinomial Logistic Regession - 
                Instead of breaking the problem into multiple binary classifications, it directly predicts
                probabilities for all classes using the softmax function.
       
"""
```

```python
#Q12 Write a Python program to train a Logistic Regression model and evaluate its performance using Precision, 
#    Recall, and F1-Score
```

```python
X,y = make_classification(n_samples= 1000, n_features= 10, n_redundant= 5, n_informative= 5, n_classes= 2, random_state= 1)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)[:,1]
```

```python
y_pred_proba
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
thresholds = np.linspace(0,1,100)
precisions = []
recalls = []
f1s = []


for threshold in thresholds:
    y_pred_threshold = (y_pred_proba >= threshold).astype(int)
    precision = precision_score(y_test, y_pred_threshold)
    recall = recall_score(y_test, y_pred_threshold)
    f1 = f1_score(y_test, y_pred_threshold)
    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)
```

```python
plt.plot(thresholds, precisions, label = 'Precision')
plt.plot(thresholds,recalls, label = 'Recall')
plt.plot(thresholds,f1s, label = 'F1')
plt.title("Precision, Recall, F1-score Vs Threshold Probability")
plt.legend()
plt.show()
```

```python
#Q24. Write a Python program to train Logistic Regression and find the optimal C (regularization strength) using 
#     cross-validation
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5, n_classes=2, random_state=1)
```

```python
#split the data into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#different C values
C_values = np.logspace(-2,2,5)
best_c = None
best_score = 0
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for C in C_values:
    model = LogisticRegression(C = C, solver= 'liblinear')
    score = np.mean(cross_val_score(model, X_train,y_train, cv = 5))
    print(f"C = {C:.2f}, Accuracy = {score:.2f}")

    if score > best_score:
        best_c = C
        best_score = score
```

```python
print(f"best C value is: {best_c}, Best Accuracy: {best_score:.2f}")
```
